### Anova is trypically used when dependent variable is continous and independent variables are categorical. By applying ANOVA to different groups, you can explore potential relationships between categorical variables and a continuous outcome. This can highlight which categorical features may be more relevant to your target variable.

```python
from scipy import stats
target_variable = 'SalePrice'

# Get categorical variables
categorical_df = train_df.select_dtypes(include=['object'])

# Initialize a dictionary to store ANOVA results
anova_results = {}

# Perform ANOVA for each categorical variable
for cat_var in categorical_df.columns:
    # Group the data by the categorical variable and collect the target variable
    grouped_data = [group[target_variable].values for name, group in train_df.groupby(cat_var)]
    
    # Perform the ANOVA test
    f_stat, p_value = stats.f_oneway(*grouped_data)
    
    # Store the results
    anova_results[cat_var] = {'F-statistic': f_stat, 'p-value': p_value}

# Convert results to DataFrame for better readability
anova_results_df = pd.DataFrame(anova_results).T
print(anova_results_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer

# Step 1: Select numeric and categorical columns
def select_columns(df):
    numeric_cols = df.select_dtypes(include=['int64', 'float64'])
    categorical_cols = df.select_dtypes(include=['object'])
    return numeric_cols, categorical_cols

# Step 2: Custom function to impute using the median of nearest neighbors for numeric data
def knn_median_impute(numeric_df, n_neighbors=3):
    # Use KNNImputer to get the nearest neighbors for numeric columns
    imputer = KNNImputer(n_neighbors=n_neighbors)
    imputed_array = imputer.fit_transform(numeric_df)

    # Create a DataFrame to hold the imputed values
    imputed_df = pd.DataFrame(imputed_array, columns=numeric_df.columns)
    
    # Loop over each column to replace missing values with the median of K-neighbors
    for i, col in enumerate(numeric_df.columns):
        if numeric_df[col].isnull().any():  # Only apply to columns with missing values
            # Get the indices of missing values
            missing_idx = numeric_df[numeric_df[col].isnull()].index
            
            for idx in missing_idx:
                # Get the K nearest neighbors' values for this missing index
                neighbors = imputed_array[:, i]
                
                # Ignore any NaNs in the neighbors and calculate the median
                k_neighbors = neighbors[~np.isnan(neighbors)]
                
                # Impute using the median of the K nearest neighbors
                median_value = np.median(k_neighbors)
                imputed_df.loc[idx, col] = median_value

    # Return the imputed numeric DataFrame
    return imputed_df

# Step 3: Function to impute missing values in categorical columns using the mode
def mode_impute(categorical_df):
    # Iterate over each categorical column and fill NaN with the mode
    for col in categorical_df.columns:
        if categorical_df[col].isnull().any():  # Only apply to columns with missing values
            mode_value = categorical_df[col].mode()[0]  # Find the mode
            categorical_df[col].fillna(mode_value, inplace=True)  # Fill missing values with mode
    return categorical_df

# Step 4: Impute both train_df and test_df
def impute_dataset(df):
    numeric_cols, categorical_cols = select_columns(df)
    
    # Apply KNN median imputation to numeric columns
    numeric_df = knn_median_impute(numeric_cols)
    
    # Apply mode imputation to categorical columns
    categorical_df = mode_impute(categorical_cols)

    # Update the original dataframe with the imputed values
    df.update(numeric_df)
    df.update(categorical_df)

    return df

# Apply to train_df
train_df = impute_dataset(train_df)

# Apply to test_df
test_df = impute_dataset(test_df)


```

```python
import pandas as pd
import numpy as np

# Feature engineering for both train and test data

# ---- Age of the house ----
train_df['Age'] = train_df['YrSold'] - train_df['YearBuilt']
test_df['Age'] = test_df['YrSold'] - test_df['YearBuilt']

# ---- Total number of bathrooms ----
train_df['TotalBath'] = train_df['FullBath'] + 0.5 * train_df['HalfBath'] + train_df['BsmtFullBath'] + 0.5 * train_df['BsmtHalfBath']
test_df['TotalBath'] = test_df['FullBath'] + 0.5 * test_df['HalfBath'] + test_df['BsmtFullBath'] + 0.5 * test_df['BsmtHalfBath']

# ---- Total number of rooms ----
train_df['TotalRooms'] = train_df['TotRmsAbvGrd'] + train_df['BedroomAbvGr']
test_df['TotalRooms'] = test_df['TotRmsAbvGrd'] + test_df['BedroomAbvGr']

# ---- Total floor square footage ----
train_df['TotalFlrSF'] = train_df['1stFlrSF'] + train_df['2ndFlrSF'] + train_df['TotalBsmtSF']
test_df['TotalFlrSF'] = test_df['1stFlrSF'] + test_df['2ndFlrSF'] + test_df['TotalBsmtSF']

# ---- Total outdoor square footage ----
train_df['TotalOutdoorSF'] = train_df['WoodDeckSF'] + train_df['OpenPorchSF'] + train_df['EnclosedPorch'] + train_df['3SsnPorch'] + train_df['ScreenPorch']
test_df['TotalOutdoorSF'] = test_df['WoodDeckSF'] + test_df['OpenPorchSF'] + test_df['EnclosedPorch'] + test_df['3SsnPorch'] + test_df['ScreenPorch']

# ---- Garage size (combining cars and area) ----
train_df['GarageSize'] = train_df['GarageCars'] * train_df['GarageArea']
test_df['GarageSize'] = test_df['GarageCars'] * test_df['GarageArea']

# ---- Lot frontage bins ----
train_df['LotFrontage_Bin'] = pd.qcut(train_df['LotFrontage'], 3, labels=['Low', 'Medium', 'High'])
test_df['LotFrontage_Bin'] = pd.qcut(test_df['LotFrontage'], 3, labels=['Low', 'Medium', 'High'])

# ---- Price per square foot ----
# train_df['PricePerSqFt'] = train_df['SalePrice'] / train_df['TotalFlrSF']
# In the test set, since 'SalePrice' is unknown, this feature would only be used during training

# ---- Remodeled indicator ----
train_df['Remodeled'] = (train_df['YearBuilt'] != train_df['YearRemodAdd']).astype(int)
test_df['Remodeled'] = (test_df['YearBuilt'] != test_df['YearRemodAdd']).astype(int)

# ---- Time since remodel ----
train_df['YearsSinceRemodel'] = train_df['YrSold'] - train_df['YearRemodAdd']
test_df['YearsSinceRemodel'] = test_df['YrSold'] - test_df['YearRemodAdd']

# ---- Garage age ----
train_df['GarageAge'] = train_df['YrSold'] - train_df['GarageYrBlt']
test_df['GarageAge'] = test_df['YrSold'] - test_df['GarageYrBlt']

# ---- Regular lot shape ----
train_df['IsRegularLotShape'] = train_df['LotShape'].apply(lambda x: 1 if x == 'Reg' else 0)
test_df['IsRegularLotShape'] = test_df['LotShape'].apply(lambda x: 1 if x == 'Reg' else 0)

# ---- Neighborhood quality (mean sale price by neighborhood) ----
neighborhood_quality = train_df.groupby('Neighborhood')['SalePrice'].mean().to_dict()
train_df['NeighborhoodQuality'] = train_df['Neighborhood'].map(neighborhood_quality)
test_df['NeighborhoodQuality'] = test_df['Neighborhood'].map(neighborhood_quality)

# ---- Basement presence ----
train_df['HasBasement'] = (train_df['TotalBsmtSF'] > 0).astype(int)
test_df['HasBasement'] = (test_df['TotalBsmtSF'] > 0).astype(int)

# ---- Total porch square footage ----
train_df['TotalPorchSF'] = train_df['OpenPorchSF'] + train_df['EnclosedPorch'] + train_df['3SsnPorch'] + train_df['ScreenPorch']
test_df['TotalPorchSF'] = test_df['OpenPorchSF'] + test_df['EnclosedPorch'] + test_df['3SsnPorch'] + test_df['ScreenPorch']

# ---- Interaction: Overall quality x living area ----
train_df['Qual_LivArea'] = train_df['OverallQual'] * train_df['GrLivArea']
test_df['Qual_LivArea'] = test_df['OverallQual'] * test_df['GrLivArea']

# ---- Seasonality of sale ----
train_df['IsHighSeason'] = train_df['MoSold'].apply(lambda x: 1 if x in [4, 5, 6, 7, 8] else 0)
test_df['IsHighSeason'] = test_df['MoSold'].apply(lambda x: 1 if x in [4, 5, 6, 7, 8] else 0)

# Now the dataset has additional engineered features ready for model training!
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer, r2_score
import numpy as np

# Step 1: Initialize the RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Step 2: Define the number of folds for cross-validation
cv_folds = 5

# Step 3: Perform cross-validation for MSE
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)
mse_scores = cross_val_score(rf_regressor, train_df_v1, target, cv=cv_folds, scoring=mse_scorer)

# Step 4: Calculate and print the average MSE across folds
mean_mse = -np.mean(mse_scores)
print(f"Cross-validated Mean Squared Error (MSE): {mean_mse:.2f}")
print(f"MSE for each fold: {-mse_scores}")

# Step 5: Perform cross-validation for R^2 score
r2_scores = cross_val_score(rf_regressor, train_df_v1, target, cv=cv_folds, scoring='r2')
mean_r2 = np.mean(r2_scores)
print(f"Cross-validated R^2: {mean_r2:.2f}")
print(f"R^2 for each fold: {r2_scores}")
```

```python
# Import necessary libraries
from skopt import BayesSearchCV
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ---- Random Forest Regressor ----

# Instantiate the RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Define the hyperparameter search space for Random Forest
param_space_rf = {
    'n_estimators': (100, 500),  # Range of values
    'max_depth': (10, 30),
    'min_samples_split': (2, 10),
    'min_samples_leaf': (1, 4),
    'max_features': [ 'sqrt', 'log2']
}

# Perform Bayesian Search with 5-fold cross-validation for Random Forest
bayes_search_rf = BayesSearchCV(estimator=rf, search_spaces=param_space_rf, n_iter=32, cv=5, n_jobs=-1, verbose=2)

# Fit the search to the data
bayes_search_rf.fit(train_df_v1, target)

# Best hyperparameters for Random Forest
print("Best Parameters for Random Forest:", bayes_search_rf.best_params_)

# Evaluate the model with the best found parameters
best_rf = bayes_search_rf.best_estimator_

# Predict and evaluate Random Forest
y_pred_rf = best_rf.predict(train_df_v1)
mse_rf = mean_squared_error(target, y_pred_rf)
r2_rf = r2_score(target, y_pred_rf)
print("Random Forest MSE:", mse_rf, "R²:", r2_rf)

# ---- Gradient Boosting Regressor ----

# Instantiate the GradientBoostingRegressor
gb = GradientBoostingRegressor(random_state=42)

# Define the hyperparameter search space for Gradient Boosting
param_space_gb = {
    'n_estimators': (100, 500),
    'learning_rate': (0.01, 0.1, 'log-uniform'),
    'max_depth': (3, 5),
    'min_samples_split': (2, 10),
    'min_samples_leaf': (1, 4)
}

# Perform Bayesian Search with 5-fold cross-validation for Gradient Boosting
bayes_search_gb = BayesSearchCV(estimator=gb, search_spaces=param_space_gb, n_iter=32, cv=5, n_jobs=-1, verbose=2)

# Fit the search to the data
bayes_search_gb.fit(train_df_v1, target)

# Best hyperparameters for Gradient Boosting
print("Best Parameters for Gradient Boosting:", bayes_search_gb.best_params_)

# Evaluate the model with the best found parameters
best_gb = bayes_search_gb.best_estimator_
y_pred_gb = best_gb.predict(train_df_v1)
mse_gb = mean_squared_error(target, y_pred_gb)
r2_gb = r2_score(target, y_pred_gb)
print("Gradient Boosting MSE:", mse_gb, "R²:", r2_gb)

# ---- XGBoost Regressor ----

# Instantiate the XGBoost Regressor
xgboost = xgb.XGBRegressor(random_state=42)

# Define the hyperparameter search space for XGBoost
param_space_xgb = {
    'n_estimators': (100, 500),
    'max_depth': (3, 6),
    'learning_rate': (0.01, 0.1, 'log-uniform'),
    'min_child_weight': (1, 3),
    'subsample': (0.6, 1.0),
    'colsample_bytree': (0.6, 1.0),
    'reg_alpha': (0, 1.0),  # L1 regularization
    'reg_lambda': (1.0, 10.0)  # L2 regularization
}

# Perform Bayesian Search with 5-fold cross-validation for XGBoost
bayes_search_xgb = BayesSearchCV(estimator=xgboost, search_spaces=param_space_xgb, n_iter=32, cv=5, n_jobs=-1, verbose=2)

# Fit the search to the data
bayes_search_xgb.fit(train_df_v1, target)

# Best hyperparameters for XGBoost
print("Best Parameters for XGBoost:", bayes_search_xgb.best_params_)

# Evaluate the model with the best found parameters
best_xgb = bayes_search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(train_df_v1)
mse_xgb = mean_squared_error(target, y_pred_xgb)
r2_xgb = r2_score(target, y_pred_xgb)
print("XGBoost MSE:", mse_xgb, "R²:", r2_xgb)
```
